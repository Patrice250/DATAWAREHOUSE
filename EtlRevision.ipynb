# Extract data

```python
import pandas as pd
```

```python
df = pd.read_excel("financial_sample.xlsx")
df
```

```python
df1 = pd.read_excel("Electric_Vehicle_Population_Data.xlsx")
df1
```

```python
Date = df[['Month Number', 'Month Name', 'Year']]
Date = Date.rename(columns={'Month Number': 'd_date', 'Month Name': 'd_month','Year': 'd_year'})
Date 
```

```python
Sales = df[['Gross Sales','Sale Price']]
Sales = Sales.rename(columns={'Gross Sales': 'quantity', 'Sale Price': 'price'})
Sales
```

```python
Sales["total_sales"] = Sales["quantity"] * Sales["price"]
Sales
```

```python
Product = df1[['Electric Vehicle Type', 'Model']]
Product = Product.rename(columns={'Electric Vehicle Type': 'P_Name', 'Model':'category'})
Product
```

```python
Customer = df[['Segment', 'Country', ]]
Customer = Customer.rename(columns={'Segment': 'c_name', 'Country': 'city'})
Customer
```

```python
import mysql.connector
mydb = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="etl_db"
)
```

```python
cursor = mydb.cursor()
```

```python
Date = Date.fillna('')
Date
```

```python
insert_query = f"INSERT INTO date_tb ({','.join(Date.columns)}) VALUES ({','.join(['%s']*len(Date.columns))})"
    
for _, row in Date.iterrows():
    values = tuple(row)
    cursor.execute(insert_query, values)
    
mydb.commit()
```

```python
Customer = Customer.fillna('')
Customer
```

```python
# Execute SQL query
cursor.execute("SELECT * FROM product_tb")
# Fetch data and column names
rows = cursor.fetchall()
column_names = [i[0] for i in cursor.description]

# Create DataFrame with column names
ProductData = pd.DataFrame(rows, columns=column_names)

# Display results
ProductData
```

```python
# Execute SQL query
cursor.execute("SELECT * FROM customer_tb")
# Fetch data and column names
rows = cursor.fetchall()
column_names = [i[0] for i in cursor.description]

# Create DataFrame with column names
CustomerData = pd.DataFrame(rows, columns=column_names)

# Display results
CustomerData
```

```python
# Execute SQL query
cursor.execute("SELECT * FROM date_tb")
# Fetch data and column names
rows = cursor.fetchall()
column_names = [i[0] for i in cursor.description]

# Create DataFrame with column names
DateData = pd.DataFrame(rows, columns=column_names)

# Display results
DateData
```

```python
date_cols = DateData[['d_id']]
customer_cols = CustomerData[['c_id']]
product_cols = ProductData[['p_id']]
sales_cols = Sales[['quantity','price']]
```
